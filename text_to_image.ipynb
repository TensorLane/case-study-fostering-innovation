## Text to Image

```python
from diffusers import AutoPipelineForText2Image
import torch

text_2_img_pipe = AutoPipelineForText2Image.from_pretrained("stabilityai/sdxl-turbo")
text_2_img_pipe.to("cpu")

prompt = "A cat warrior going against t-rex."

image = text_2_img_pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]
image
```

```python
prompt = "A small cat warrior going against huge t-rex in mountain"

image = text_2_img_pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]
image
```

## Image to Image

```python
from diffusers import AutoPipelineForImage2Image
from diffusers.utils import load_image
import torch

img_2_img_pipe = AutoPipelineForImage2Image.from_pretrained("stabilityai/sdxl-turbo")
img_2_img_pipe.to("cpu")

init_image = load_image("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/cat.png").resize((512, 512))

prompt = "cat warrior, sword, wings, detailed, fantasy, cool, powerful, Pixar, 8k"

image = img_2_img_pipe(prompt, image=init_image, num_inference_steps=2, strength=0.5, guidance_scale=0.0).images[0]
image
```

```python
prompt = "cat warrior, sword, gold plated armor, golden helmet, wings, detailed, fantasy, cool, powerful, Pixar, 8k"

image = img_2_img_pipe(prompt, image=init_image, num_inference_steps=2, strength=0.5, guidance_scale=0.0).images[0]
image
```
